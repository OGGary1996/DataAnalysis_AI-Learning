## 4. 数据前置检查和处理

##### 1. 导入数据

```python
import pandas as pd

df = pd.read_csv('../../02 Programming for Data Science/LO3/autompg.csv')
```

#### 2. 数据预览
查看数据的前几行和后几行，了解数据的基本结构和内容。

#### 3. 查看数据的基本信息
使用 .info() 方法查看数据的基本信息，包括数据类型、非空值数量等。<br>
使用场景：
1. 哪些列存在 Nan 缺失值
2. 每列的数据类型是否正确

```python
df.info()
```

#### 4. 统计描述
使用 .describe() 方法查看数值型数据的统计描述信息，如均值、标准差、最小值、最大值等。<br>
使用场景：
1. 快速发现数据的异常值 outlier
2. 判断数据分布是否合理，如均值和中位数的差异
3. 是否存在明显的错误数据，比如负值等和极大值

```python
df.describe()
```

#### 5. 查看列明和数据维度
使用 .columns 属性查看数据的列名，使用 .shape 属性查看数据的维度（行数和列数）。<br>
使用场景：
1. 了解数据的结构
2. 确认数据的行列数量是否符合预期

```python
df.shape
```

```python
df.columns
```

#### 6. 检查数据类型
使用 .dtypes 属性查看每列的数据类型。<br>
使用场景：
1. 确认每列的数据类型是否正确
2. 发现数据类型不匹配的问题，比如数值型数据被错误地识别为字符串类型

```python
df.dtypes
```

#### 7. 检查缺失值
使用 .isna()方法检查每列的缺失值数量；返回值为布尔值的 DataFrame，表示每个元素是否为缺失值。<br>
使用 .isna().sum() 可以统计每列的缺失值数量。<br>
使用场景：
1. 了解数据中缺失值的分布情况
2. 决定是否需要进行缺失值处理，如填充或删除

```python
df.isna().sum()
```

#### 8. 检查重复值
使用 .duplicated() 方法检查数据中的重复行数量；返回值为布尔值的 Series，表示每行是否为重复行。<br>
使用 .duplicated().sum() 可以统计重复行的数量。<br>
使用场景：
1. 了解数据中重复值的情况
2. 决定是否需要删除重复值
3. 确保数据的唯一性

```python
df.duplicated().sum()
```
