## 2. Series 的常用属性

```python
import numpy as np
import pandas as pd
from numpy.ma.core import size
```

#### 1. 常用属性

```python
dict_age = {
    "Tom": 18,
    "Jack": 20,
    "Steve": 25,
    "Ricky": 30,
    "Anna": 22
}
s_age = pd.Series(dict_age, dtype=np.int32, name="Age")

# 常用属性
print(s_age.index)
print(s_age.values)
print(s_age.dtype)
print(s_age.name)
print(s_age.size)
print(s_age.shape)
```

#### 2. 与 Nan 相关的属性

```python
arr = [1,2,np.nan,4,np.nan,6]
s_nan = pd.Series(arr, dtype=np.float32, name="With NaN")
print(s_nan.hasnans)

# 返回 bool 组成的 Series，表示每个位置是否为 NaN
print(s_nan.isna())
print(s_nan.notna())
```
