#### 2. 数据类型转换核心方法

##### 2.1 使用 astype() 方法转换数据类型
注意：
1. 转换为数值类型时，如果存在无法转换的值，会抛出错误。
2. 不适合直接对脏数据进行转换。

##### 2.2 使用 to_numeric() 方法转换为数值类型

```python
pd.to_numeric(df['hp'], errors='coerce')
```

##### 2.3 使用 to_datetime() 方法转换为日期时间类型

#### 3. 处理字符串格式
使用 .str 访问器，可以对字符串列进行各种操作。

##### 3.1 去除字符串的空白字符

```python
df["name"] = df["name"].str.strip()
```

##### 3.2 字符串的大小写转换

```python
df["name"] = df["name"].str.lower()
```

##### 3.3 特殊字符的替换

```python
df["name"] = df["name"].str.replace(",", " ")
```
