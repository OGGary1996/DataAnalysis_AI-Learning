## 5.2 数据清洗 - 处理重复值

```python
import pandas as pd

df = pd.read_csv('../../02 Programming for Data Science/LO3/autompg.csv')
df.head()
```

#### 1. 检测重复值
使用 duplicated() 方法，返回一个布尔值的 Series，表示每行是否为重复值。

```python
df.duplicated()
```

```python
df.duplicated().sum()
```

#### 2. 删除重复值
使用 drop_duplicates() 方法，删除重复行。
方法参数：
1. subset: 指定用于检测重复值的列，默认使用所有列，指定了列后，只会考虑这些列的值来判断是否重复。
2. keep: 指定保留哪一行，'first'（默认值）保留第一次出现的行，'last' 保留最后一次出现的行，False 删除所有重复行。
3. inplace: 是否在原 DataFrame 上进行修改，默认 False。

```python
df.drop_duplicates(
    subset=["mpg", "cyl", "displ", "hp"],
    keep='first',
    inplace=False
)
```
