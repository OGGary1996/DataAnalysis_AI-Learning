## 3. Series 的索引与定位

```python
import numpy as np
import pandas as pd
```

```python
s = pd.Series(
    [18, 20, 25, 30, 22],
    index=["Tom", "Jack", "Steve", "Ricky", "Anna"],
    dtype=np.int32,
    name="Age of Employees"
)
print(s)
print(s.index)
print(s.values)
print(s.name)
print(s.dtype)
print(s.shape)
print(s.size)
```

#### 1. 通过原始的写法 [] 进行定位

```python
print("Age at index 2: ", s[2]) # 通过位置索引定位
print("Age of Tom: ", s["Tom"]) # 通过标签索引定位
# 注意：如果此时标签是数字标签，即0,1,2,...，则会产生歧义，此时会优先按照位置索引进行定位
```

#### 2. 通过 .loc 进行标签索引定位

```python
# 取单个值
print("Age of Jack: ", s.loc["Jack"])
# 取多个值
print("Age of Steve and Anna: ", s.loc[["Steve", "Anna"]])
# 取连续的多个值（切片）
print("Age from Tom to Ricky: ", s.loc["Tom": "Ricky"])
```

#### 3. 通过 .iloc 进行位置索引定位

```python
# 取单个值
print("Age at index 3: ", s.iloc[3])
# 取多个值
print("Age at index 1 and 4: ", s.iloc[[0,2]])
# 取连续的多个值（切片）
print("Age from index 1 to 3: ", s.iloc[1:4])
```

#### 4. 通过布尔索引进行定位

```python
s_gt_25 = s[s >= 25] # 取出年龄大于等于25的员工
print("Employees with age >= 25:\n", s_gt_25)
```
