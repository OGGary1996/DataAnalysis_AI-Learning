## 4. 散点图

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 6]
```

#### 1. 基本的散点图

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
plt.show()
```

#### 2. 样式优化
1. s/size : 点的大小
2. c/color : 点的颜色
3. marker : 点的形状
4. alpha : 点的透明度

```python
fig, ax = plt.subplots()
ax.scatter(x,y, s=90, c='orange', marker='^', alpha=0.7)
plt.show()
```

##### 注意：
1. c 可以通过传入一个颜色列表来实现每个点不同颜色，或者渐变色
 - c = x/y 表示颜色根据 x/y 值变化，cmap 参数可以指定颜色映射表，如 cmap='viridis'，表示使用 Viridis 颜色映射表进行颜色渐变
2. s 也可以通过传入一个大小列表来实现每个点不同大小
 - s = x/y 表示大小根据 x/y 值变化

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s=y, c=y, cmap='viridis', alpha=0.7)
plt.show()
```

#### 3. 添加标题和标签

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s=90, c=y, cmap='viridis', alpha=0.7, marker='o')
ax.set_title('Scatter plot')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
plt.show()
```

#### 4. 网格与显示范围

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s=90, c=y, cmap='viridis', alpha=0.7, marker='o')
ax.set_title('Scatter plot')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
plt.show()
```
