## OO 绘画风格

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.DataFrame({
    "Age": [25, 30, 35, 40, 45],
    "Salary": [50000, 60000, 70000, 80000, 90000]
})
```

#### 1. Matplotlib 的结构

matplotlib 是一个非常强大的绘图库，它的设计基于面向对象的编程（OOP）理念。理解 matplotlib 的结构对于有效地使用它非常重要。
以下是 matplotlib 的主要组件：
1. **Figure（图形）**：Figure 是整个绘图的容器。它可以包含多个子图（Axes）。
2. **Axes（坐标轴）**：Axes 是 Figure 中的一个区域，用于绘制数据。每个 Axes 可以包含多个图形元素（如线条、点等）。
3. **Axis（轴）**：Axis 是 Axes 的一部分，用于表示数据的刻度和标签。每个 Axes 通常有两个 Axis（x 轴和 y 轴）。
4. **Artist（艺术家）**：Artist 是 matplotlib 中的基本绘图元素。所有可见的东西（如线条、文本、图形等）都是 Artist 的子类。

#### 2. 2种绘图风格

##### 2.1 命令式风格（Pyplot 风格）
特点：
1. 直接使用 plt.plot() 等函数进行绘图。
2. 此时会自动创建 Figure 和 Axes 对象。
3. 适合快速绘图和简单的可视化任务。

```python
# 自动创建 Figure 和 Axes
plt.plot(df["Age"], df["Salary"])
# 使用 plt. 命令修改属性
plt.title("Salary vs Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

##### 2.2 面向对象风格（OO 风格）
特点：
1. 显式创建 Figure 和 Axes 对象。
2. 通过 Figure 和 Axes 对象的方法进行绘图和修改属性。
3. 适合复杂的绘图任务和需要高度自定义的可视化

```python
# 显式创建 Figure 和 Axes
# subplots 表示创建多个子图，但是这里只有一个子图
fig, ax = plt.subplots()
# 使用 fig 和 ax 对象的方法进行绘图和修改属性
fig.set_size_inches(7, 5)
ax.plot(df["Age"], df["Salary"])
ax.set_title("Salary vs Age")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
plt.show()
```
