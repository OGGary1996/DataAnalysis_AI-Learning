## 3. 柱状图(Bar Chart)

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [23, 45, 12, 37]
```

#### 1. 基础的柱状图

##### 1.1 竖向柱状图

```python
fig, ax = plt.subplots()
ax.bar(categories, values)
ax.set_title("Basic Bar Chart")
plt.show()
```

##### 1.2 横向柱状图
使用 ax.barh() 方法可以创建横向柱状图。

```python
fig, ax = plt.subplots()
ax.barh(categories, values)
ax.set_title("Basic Horizontal Bar Chart")
plt.show()
```

#### 2. 自定义样式
1. color: 设置柱子的颜色
2. edgecolor: 设置柱子边框颜色
3. width: 设置柱子的宽度
4. alpha: 设置柱子的透明度
5. label: 为柱子添加图例标签

```python
fig, ax = plt.subplots()
ax.bar(categories, values, color='lightgreen', edgecolor='lightblue', width=0.5, alpha=0.7)
ax.set_title("Customized Bar Chart")
plt.show()
```

#### 3. 坐标轴和标题
1. 设置坐标轴标签, ax.set_xlabel() 和 ax.set_ylabel()
2. 设置标题, ax.set_title()

```python
fig, ax = plt.subplots()
ax.bar(categories, values, color='lightgreen', edgecolor='lightblue', width=0.5, alpha=0.7)
ax.set_title("Bar Chart with Axis Labels and Grid")
ax.set_xlabel("Categories")
ax.set_ylabel("Values")
plt.show()
```

#### 4. 网格和y轴显示范围控制
1. 添加网格, ax.grid(True)， 如果要添加y轴网格，可以使用参数 axis='y'，例如 ax.grid(True, axis='y')
2. 设置y轴范围, ax.set_ylim()，参数为一个元组，表示y轴的最小值和最大值

```python
fig, ax = plt.subplots()
ax.bar(categories, values, color='purple', edgecolor='blue', width=0.7, alpha=0.7)
ax.set_title("Bar Chart with Grid and Y-axis Limits")
ax.set_xlabel("Categories")
ax.set_ylabel("Values")
ax.grid(True, axis='y', linestyle='--', linewidth=0.5)
ax.set_ylim(0, 60)
plt.show()
```

#### 5. 添加数据标签
每个柱子在ax中都是一个Rectangle对象，可以通过ax.patches获取所有柱子的列表，然后使用ax.text()方法在每个柱子上方添加数据标签。

```python
fig, ax = plt.subplots()
ax.bar(categories, values, color='lightgreen', edgecolor='lightblue', width=0.75, alpha=0.7)
ax.set_title("Bar Chart with Data Labels")
ax.set_xlabel("Categories")
ax.set_ylabel("Values")
ax.grid(axis='y', linestyle='--', linewidth=0.5)
ax.set_ylim(0,60)
for bar in ax.patches:
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2, # x坐标
        height,                 # y坐标
        f'{height}',           # 标签内容
        ha='center',           # 水平对齐方式
        va='bottom'            # 垂直对齐方式
    )
plt.show()
```
