## 2. 线图(Line Chart)

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y_1 = [10, 15, 13, 17, 20]
y_2 = [12, 14, 16, 18, 22]
y_3 = [14, 13, 15, 19, 21]
```

#### 1. 最基础的线图

```python
fig, ax = plt.subplots()
ax.plot(x, y_1)
ax.plot(x, y_2)
ax.plot(x, y_3)
ax.set_title("Basic Line Chart")
plt.show()
```

#### 2. 修改优化样式：
1. 对于多个线，必须要添加图例（legend）
2. 修改线条样式, linestyle 参数
3. 修改线条颜色, color 参数
4. 修改线条宽度, linewidth 参数
5. 添加数据点标记, marker 参数

```python
fig, ax = plt.subplots()
ax.plot(x, y_1, label="first line", linestyle="-", color="red", marker="o", linewidth=2)
ax.plot(x, y_2, label="second line", linestyle="--", color="blue", marker="s", linewidth=2)
ax.plot(x, y_3, label="third line", linestyle=":", color="green", marker="^", linewidth=2)
ax.set_title("Customized Line Chart")
ax.legend()
plt.show()
```

#### 3. 其他补充优化
1. 添加网格, ax.grid(True)
2. 设置坐标轴范围, ax.set_xlim() 和 ax.set_ylim()，适用于数据范围较大时放大查看细节
3. 添加注释, ax.annotate()

```python
fig, ax = plt.subplots()
ax.plot(x, y_1, label="first line", linestyle="-", color="red", marker="o", linewidth=2)
ax.plot(x, y_2, label="second line", linestyle="--", color="blue", marker="s", linewidth=2)
ax.plot(x, y_3, label="third line", linestyle=":", color="green", marker="^", linewidth=2)
ax.set_title("Customized Line Chart")
ax.legend()
ax.set_title("Customized Line Chart with Grid and Annotations")
ax.grid(True)
ax.set_xlim(1, 3)
ax.set_ylim(10, 18)
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```
