## 7. Seaborn

Seaborn 是一个基于 Matplotlib 的数据可视化库，提供了更高级的接口和美观的默认样式，适合进行统计图表的绘制。它简化了复杂图表的创建过程，并且与 Pandas 数据结构紧密集成，使得数据可视化更加便捷。

```python
# conda install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'A': np.random.rand(100),
    'B': np.random.rand(100),
    'C': np.random.choice(['X', 'Y', 'Z'], size=100)
})
df
```

#### 1. 创建关系型图表

##### 1.1 散点图 (Scatter Plot)
用于快速查看两个变量之间的关系。

```python
fig, ax = plt.subplots()
sns.scatterplot(
    data=df,
    x="A",
    y="B",
    ax=ax, # 指定绘图的坐标轴，来自于 Matplotlib
    hue="C" # 根据类别变量 C 着色
)
plt.show()
```

##### 1.2 折线图 (Line Plot)
用于显示数据点随时间或其他连续变量的变化趋势。

```python
fig, ax = plt.subplots()
sns.lineplot(
    data=df,
    x="A",
    y="B",
    ax=ax,
    hue="C",
    estimator=None # 不进行聚合，显示所有数据点
)
plt.show()
```

#### 2. 分类图表

##### 2.1 条形图 (Bar Plot)
用于比较不同类别的数值。

```python
fig, ax = plt.subplots()
sns.barplot(
    data=df,
    x="C",
    y="A",
    ax=ax
)
plt.show()
```

##### 2.2 箱线图 (Box Plot)
用于显示数据的分布情况和异常值。

```python
fig, ax = plt.subplots()
sns.boxplot(
    data=df,
    x="C",
    y="A",
    ax=ax
)
plt.show()
```

##### 2.3 小提琴图 (Violin Plot)
结合了箱线图和核密度估计，显示数据的分布形态

```python
fig, ax = plt.subplots()
sns.violinplot(
    data=df,
    x="C",
    y="A",
    ax=ax
)
plt.show()
```

#### 3. 单变量分布图

##### 3.1 直方图 (Histogram)
用于显示单个变量的分布情况。

```python
fig, ax = plt.subplots()
sns.histplot(
    data=df,
    x="A",
    ax=ax,
    bins=10,
    kde=True # 添加核密度估计曲线
)
plt.show()
```

##### 3.2 KDE 图 (Kernel Density Estimate Plot)
用于显示单个变量的概率密度分布。

```python
fig, ax = plt.subplots()
sns.kdeplot(
    data=df,
    x="A",
    ax=ax,
    fill=True # 填充曲线下方区域
)
plt.show()
```

#### 4. 矩阵图表

##### 4.1 相关系数热图 (Correlation Heatmap)

```python
corr = df[["A", "B"]].corr() # 计算数值列之间的相关系数
fig, ax = plt.subplots()
sns.heatmap(
    data=corr,
    ax=ax,
    annot=True, # 显示数值
)
plt.show()
```

##### 4.2 缺失值热图 (Missing Values Heatmap)

```python
fig, ax = plt.subplots()
sns.heatmap(
    data=df.isnull(),
    ax=ax,
    cbar=False, # 不显示颜色条
)
plt.show()
```
