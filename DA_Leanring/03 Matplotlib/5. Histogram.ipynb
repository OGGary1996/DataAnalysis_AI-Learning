## 5. 直方图 (Histogram)

```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import alpha

data = pd.Series([12, 15, 14, 10, 18, 20, 22, 25, 30, 28, 35, 40, 38, 45, 50, 55, 60, 65, 70, 75])
```

#### 1. 基础直方图

```python
fig, ax = plt.subplots()
ax.hist(data, bins=5)
ax.set_title("Basic Histogram")
plt.show()
```

#### 2. 自定义样式
1. color: 设置柱子的颜色
2. edgecolor: 设置柱子边框颜色
3. alpha: 设置柱子的透明度
4. bins: 设置柱子的数量或边界
5. orientation: 设置柱子的方向，'vertical' 或 'horizontal'
6. label: 为柱子添加图例标签


```python
fig, ax = plt.subplots()
ax.hist(
    data,
    bins=5,
    color="purple",
    edgecolor="white",
    alpha=0.7,
    orientation="vertical",
)
ax.set_title("Histogram")
ax.set_xlabel("Value Ranges")
ax.set_ylabel("Frequency")
plt.show()
```

#### 3. 多个数据集的直方图

```python
data_1 = pd.Series([12, 15, 14, 10, 18, 20, 22, 25, 30, 28])
data_2 = pd.Series([35, 40, 38, 45, 50, 55, 60, 65, 70, 75])

fig, ax = plt.subplots()
ax.hist(
    data_1,
    color="purple",
    edgecolor="white",
    alpha=0.7,
    bins=3,
    label="Dataset 1",
)
ax.hist(
    data_2,
    color="orange",
    edgecolor="white",
    alpha=0.7,
    bins=5,
    label="Dataset 2",
)
ax.legend()
ax.set_title("Histogram")
ax.set_xlabel("Value Ranges")
ax.set_ylabel("Frequency")
plt.show()
```
