## 6. 箱线图
显示内容：
1. 数据最小值
2. 第一四分位数 (Q1)
3. 中位数 (Q2)
4. 第三四分位数 (Q3)
5. 数据最大值
6. 异常值（如果有）
Q1 - Q3 之间的范围称为四分位距 (IQR)， IQR包含了数据的中间50%。
通常，任何低于 Q1 - 1.5 * IQR 或高于 Q3 + 1.5 * IQR 的数据点都被视为异常值。箱线图通过这些统计量帮助我们理解数据的分布、集中趋势和变异性。

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.Series([12, 15, 14, 10, 18, 20, 22, 25, 30, 28, 35, 40, 38, 45, 50, 55, 60, 65, 70, 75])
```

#### 1. 基础箱线图

```python
fig, ax = plt.subplots()
ax.boxplot(data, vert=True)
ax.set_title("Basic Boxplot")
plt.show()
```

#### 2. 自定义样式
1. boxprops: 设置箱子的属性，如颜色、边框等
2. whiskerprops: 设置须的属性
3. capprops: 设置箱线图顶端和底端横线的属性
4. medianprops: 设置中位数线的属性
5. flierprops: 设置异常值点的属性

```python
fig, ax = plt.subplots()
ax.boxplot(
    data,
    vert=False, # 水平箱线图
    boxprops={"color":"lightblue", "linewidth": 2}, # 箱子边框颜色和宽度
    whiskerprops={"color":"lightgreen", "linewidth": 2}, # 须的颜色和宽度
    capprops={"color":"red", "linewidth": 2}, # 顶端和底端横线的颜色和宽度
    medianprops={"color":"orange", "linewidth": 2}, # 中位数线的颜色和宽度
    flierprops={"marker":"o", "color":"red", "alpha": 0.5}, # 异常值点的样式
)
plt.show()
```

#### 3. 多个数据集的箱线图

```python
df = pd.DataFrame({
    "Dataset 1": [12, 15, 14, 10, 18, 20, 22, 25, 30, 28],
    "Dataset 2": [35, 40, 38, 45, 50, 55, 60, 65, 70, 75],
    "Dataset 3": [20, 22, 25, 30, 28, 35, 40, 38, 45, 50],
})

fig, ax = plt.subplots()
ax.boxplot(
    [df["Dataset 1"], df["Dataset 2"], df["Dataset 3"]],
    vert=False,
    boxprops={"color":"lightblue", "linewidth": 2},
    whiskerprops={"color":"lightgreen", "linewidth": 2},
    capprops={"color":"red", "linewidth": 2},
    medianprops={"color":"orange", "linewidth": 2},
    flierprops={"marker":"o", "color":"red", "alpha": 0.5},
)
ax.set_xticklabels(['Dataset 1', 'Dataset 2', 'Dataset 3']) # 设置x轴标签,与set_xlabel不同,是设置刻度标签
ax.set_yticklabels(['Dataset 1', 'Dataset 2', 'Dataset 3'])
ax.set_title("Boxplot for Multiple Datasets")
plt.show()
```
