```python
import math

# Mathematical Constants
print(f"Pi (π): {math.pi}")
print(f"Euler's number (e): {math.e}")
print(f"Tau (τ): {math.tau}")
print(f"Positive infinity: {math.inf}")
print(f"Not a Number (NaN): {math.nan}")
```

```python
 # Number Functions
num = 4.7
print(f"ceil({num}): {math.ceil(num)}")  # 5
print(f"floor({num}): {math.floor(num)}")  # 4
print(f"factorial(5): {math.factorial(5)}")  # 120
print(f"gcd(12, 18): {math.gcd(12, 18)}")  # 6
print(f"isfinite(10): {math.isfinite(10)}")  # True
print(f"isinf(math.inf): {math.isinf(math.inf)}")  # True
print(f"isnan(math.nan): {math.isnan(math.nan)}")  # True
```

```python
#  Power and Logarithmic Functions
print(f"pow(2, 3): {math.pow(2, 3)}")  # 8.0
print(f"exp(1): {math.exp(1)}")  # e^1 ≈ 2.718
print(f"log(100, 10): {math.log(100, 10)}")  # 2.0
print(f"log2(8): {math.log2(8)}")  # 3.0
print(f"log10(1000): {math.log10(1000)}")  # 3.0
print(f"sqrt(25): {math.sqrt(25)}")  # 5.0
```

```python
# Angular Conversion
print(f"radians(180): {math.radians(180)}")  # π
print(f"degrees(math.pi): {math.degrees(math.pi)}")  # 180.0
```

```python
# Trigonometric Functions
angle_deg = 60
angle_rad = math.radians(angle_deg)
print(f"sin({angle_deg}°): {math.sin(angle_rad):.3f}")
print(f"cos({angle_deg}°): {math.cos(angle_rad):.3f}")
print(f"tan({angle_deg}°): {math.tan(angle_rad):.3f}")
print(f"hypot(3, 4): {math.hypot(3, 4)}")  # 5.0
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.title("Sine and Cosine Waves", fontsize=16)
plt.xlabel("Angle (radians)", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

x = np.linspace(0, 2 * math.pi, 500)  # 500 points for smooth curves

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, color='blue', linewidth=2, label='sin(x)')
plt.plot(x, y_cos, color='red', linewidth=2, label='cos(x)')

plt.legend(fontsize=12, loc='upper right')
plt.tight_layout()

plt.show()
```
