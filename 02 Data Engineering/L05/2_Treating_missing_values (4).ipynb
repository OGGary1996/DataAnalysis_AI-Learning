### Filling in for missing values

```python
np.random.seed(25)
DF_obj = pd.DataFrame(np.random.rand(36).reshape(6,6))
DF_obj
```

```python
DF_obj.loc[3:5, 0] = missing
DF_obj.loc[1:4, 5] = missing
DF_obj
```

```python
filled_DF = DF_obj.fillna(0)
filled_DF
```

```python
filler_DF = DF_obj.fillna({0:0.1,5:1.25})
filler_DF
```

```python
DF_obj
```

```python
filled_DF_2 = DF_obj.fillna(method='ffill')
filled_DF_2
```

```python
filled_DF_3 = DF_obj.fillna(method='bfill')
filled_DF_3
```

### Counting missing values

```python
np.random.seed(25)
DF_obj = DataFrame(np.random.rand(36).reshape(6,6))
DF_obj.loc[3:5, 0] = missing
DF_obj.loc[1:4, 5] = missing
DF_obj
```

```python
DF_obj.isnull().sum()
```

```python
DF_obj[0] = DF_obj[0].fillna(DF_obj[0].mean())
DF_obj
```
