## 3.1 使用GridSearchCV进行MNIST手写数字识别 - 数据可视化
在本节中，我们将使用GridSearchCV对KNN算法进行超参数选择，以实现对MNIST手写数字数据集的识别。首先，我们将对数据集进行可视化，以便更好地理解数据的分布和特征。

#### 1. 导入必要的库和加载MNIST数据集

```python
import numpy as np
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt

# Load MNIST dataset
X, y = fetch_openml(
    'mnist_784',
    version=1,
    return_X_y=True, # 只返回特征和标签
    as_frame=False # False 表示以NumPy数组形式返回数据，而不是Pandas DataFrame
)
print("Dataset shape:", X.shape, y.shape)  # (70000, 784) (70000,)
```

#### 2. 实现某一行的图像可视化函数

##### 2.1 定义可视化函数

```python
def plot_image(index):
    image = X[index].reshape(28,28) # 将一维数组重塑为28x28的二维数组
    label = y[index] # 获取对应的标签
    # 使用matplotlib进行可视化
    fig, ax = plt.subplots()
    ax.imshow(image, cmap='gray') # 显示图像，使用灰度色映射, 原始的MNIST图像是黑白的
    ax.set_title(f'Label: {label}') # 设置标题为标签
    ax.axis('off') # 关闭坐标轴显示
    plt.show()
```

##### 2.1 可视化示例图像

```python
plot_image(1)
```

#### 3. 可视化多张图像

```python
def plot_multiple_images(indices): # 传递索引列表，16个
    fig, axes = plt.subplots(4, 4, figsize=(8, 8)) # 创建4x4的子图
    for ax, index in zip(axes.flatten(), indices):
        image = X[index].reshape(28,28)
        label = y[index]
        ax.imshow(image, cmap='gray')
        ax.set_title(f'Label: {label}')
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
indices = np.random.choice(len(X), 16, replace=False) # 随机选择16个索引
plot_multiple_images(indices)
```
